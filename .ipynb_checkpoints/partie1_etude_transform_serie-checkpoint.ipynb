```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline
```

# Data 

```python
FolderNames = ["BoulangeriePatisserie", "GlacesSorbets", "Fromage", "Electricite"]

df = pd.read_csv(
   "./{}/valeurs_mensuelles.csv".format(FolderNames[0]),
    sep=";",
    skiprows=[1, 2],
    index_col="Libellé",
    parse_dates=["Libellé"])
df.columns = [FolderNames[0],"ToDelete"]
del(df["ToDelete"])


for i,name in enumerate(FolderNames):
    if i>0:
        temp = pd.read_csv(
       "./{}/valeurs_mensuelles.csv".format(name),
        sep=";",
        skiprows=[1, 2],
        index_col="Libellé",
        parse_dates=["Libellé"])
        temp.columns = [FolderNames[i],"ToDelete"]
        del(temp["ToDelete"])
        df = pd.merge(df, temp, on='Libellé')
    
df.head()
```

# Plots

Les indices de dates sont les mêmes, on peut donc merger les 4 séries sur les dates.

```python
fig = plt.figure(figsize=(15,15))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i,name in enumerate(FolderNames):
    
    X = df.iloc[:,i].dropna()
    ax = fig.add_subplot(4, 3, i+1)
    ax.plot(X)
    ax.set(title=name)
```

On garde la série du commerce d'électricité : tendance croissante jusqu'en 2010 et saisonalité annuelle.

# Electricité

```python
dfE = pd.read_csv(
   "./Electricite/valeurs_mensuelles.csv",
    sep=";",
    skiprows=[1, 2],
    index_col="Libellé",
    parse_dates=["Libellé"])
dfE.columns = ["Electricite","ToDelete"]
del(dfE["ToDelete"])
```

```python
plt.figure(figsize=(15,5))
plt.plot(dfE)
plt.title("Commerce mensuel d'electricite")
```

## Transformation logarithmique

```python
log_dfE = np.log(dfE)
```

```python
plt.figure(figsize=(15,5))
plt.plot(log_dfE)
plt.title("log commerce mensuel d'electricite")
```

**Commentaires:** La transformation logarithmique permet de corriger certaines formes d'hétéroscédasticité ou de non-linéarité. Dans notre cas, notre série était légèrement hétéroscédastique et la transformation logarithmique a permis de la corriger.<br><br>

On travaillera donc avec la transformation logarithmique.

## Décomposition Trend / Saisonalité / Bruit

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result_add = seasonal_decompose(log_dfE, model='additive')
result_mult = seasonal_decompose(log_dfE, model='multiplicative')

def plotseasonal(res, axes ):
    res.observed.plot(ax=axes[0], legend=False)
    axes[0].set_ylabel('Observed')
    res.trend.plot(ax=axes[1], legend=False)
    axes[1].set_ylabel('Trend')
    res.seasonal.plot(ax=axes[2], legend=False)
    axes[2].set_ylabel('Seasonal')
    res.resid.plot(ax=axes[3], legend=False)
    axes[3].set_ylabel('Residual')

fig, axes = plt.subplots(ncols=2, nrows=4, sharex=True, figsize=(12,5))

plotseasonal(result_add, axes[:,0])
plotseasonal(result_mult, axes[:,1])
plt.suptitle("Decomposition de la Prod. d'electricite")
plt.show()
```

## ACF

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
plt.figure(figsize=(10,10))
plot_acf(log_dfE.values, lags=40, title="acf log commerce Elec")
plt.show()
```

**Commentaires:** On observe une saisonalité annuelle (pic tous les 12 mois).

## PACF

```python
plt.figure(figsize=(10,10))
plot_pacf(log_dfE.values, lags=40, title="pacf log commerce Elec")
plt.show()
```

**Commentaires:** Soupcon de racine unitaire, correlation forte entre $X_t$ et $X_{t-1}$

## Désaisonnalisation 

### Première méthode : différence 12ème 

```python
log_dfE_diff=log_dfE.diff(12).dropna()
log_dfE_diff.plot(title="Log du commerce desaisonnalisee", legend=False)
```

```python
np.mean(log_dfE_diff)
```

### 2ème méthode : régression avec des "dummy"

```python
import statsmodels.api as sm
```

```python
log_dfE.shape
```

On régresse log_dfE sur les indicatrices de chaque mois : $$logdfE_t = \alpha + \sum_{i=11}^11 \beta_i \mathcal{1}_{mois_t=i} + \epsilon_t$$ $$Y = X'\beta + \epsilon$$ $X$ à spécifier (350*12)

```python
X = np.tile(np.eye(12), 32).T[:350,:11]
X = sm.add_constant(X)
X = pd.DataFrame(X,index= log_dfE.index)
X.shape
```

```python
# On fait une régression par MCO du log du commerce d'électricité sur les dummy de mois (sauf 1 pour eviter le DV trap)
model = sm.OLS(log_dfE,X) 
results = model.fit()
DeSaison = results.resid # On récupère les résidus
```

```python
print(results.summary())
```

```python
DeSaison.plot() # Reste la tendance
```

## Suppression du trend

### Méthode 1 : régression

```python
preTrend = DeSaison.loc[:"2008-12-01"]
postTrend = DeSaison.loc["2009-01-01":]
```

#### Régression pretrend

```python
X2 = np.arange(1,preTrend.shape[0]+1)
X2 = sm.add_constant(X2)
X2 = pd.DataFrame(X2,index= preTrend.index)
```

```python
model2 = sm.OLS(preTrend,X2) 
results2 = model2.fit()
preStation = results2.resid # On récupère les résidus
```

```python
print(results2.summary())
```

```python
preStation.plot() # Reste la tendance
```

#### Régression postrend

```python
X3 = np.arange(1,postTrend.shape[0]+1)
X3 = sm.add_constant(X3)
X3 = pd.DataFrame(X3,index= postTrend.index)
```

```python
model3 = sm.OLS(postTrend,X3) 
results3 = model3.fit()
postStation = results3.resid # On récupère les résidus
```

```python
print(results3.summary())
```

```python
postStation.plot() # Reste la tendance
```

#### Merge

```python
station = pd.concat([preStation,postStation])
```

```python
station.plot()
station.shape
```

### Méthode 2 : différenciation à l'ordre 1

```python
station2 = DeSaison.diff(1).dropna()
```

```python
station2.plot()
```

Semble plus stationnaire que la première méthode. À comparer avec des tests.
